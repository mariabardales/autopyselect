```python
import mysql.connector
import pandas as pd
from sqlalchemy import create_engine

def IniciarConexion(database):
    parametros = {'user': 'root',
                  'password': '****',
                  'host': 'localhost',
                  'database': f'{database}'}

    conn = mysql.connector.connect(**parametros)
    conn_cadena = f"mysql+mysqlconnector://{parametros['user']}:{parametros['password']}@{parametros['host']}/{parametros['database']}"
    engine = create_engine(conn_cadena)
    return conn,engine

def ConsultaSelect(nombre_bd,nombre_tabla):
    conexion,motor= IniciarConexion(nombre_bd)
    consulta = f"SELECT * FROM {nombre_tabla}"
    df = pd.read_sql_query(consulta, motor)
    conexion.close()
    return df

def ListaTablas(esquema):
    conexion,motor= IniciarConexion(esquema)
    cursor = conexion.cursor()
    cursor.execute("SHOW TABLES")
    resultado = cursor.fetchall()
    conexion.close()
    lista_tablas= []
    for elemento in resultado:
        lista_tablas.append(elemento[0])
    return lista_tablas

def AutoSelect(esquema):
    tablas = ListaTablas(esquema)
    lista_df=[]
    for tabla in tablas:
        df = ConsultaSelect(esquema,tabla)
        lista_df.append(df)
    return lista_df
```
